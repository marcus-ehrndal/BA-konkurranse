```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import tools
from math import ceil
from scipy.stats import zscore
from sklearn.neural_network import BernoulliRBM
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, partial_dependence
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn import preprocessing
import csv
```

```python
#folder = 'ba/'
folder = 'B:/03. Presentasjoner/Metoder og faglig/20150311 BA-konkurranse/'
file_name = folder + 'census_train.csv';
categorical_cols = ['workclass','education','marital-status','relationship',
                   'occupation','race','sex','native-country']
continuous_cols = ['age','capital-gain','capital-loss','hours-per-week']
class_col = 'salary-over-50k'
nullvalues = []
#df, e = tools.prepare(df,catcols=categorical_cols, contcols=continuous_cols, nullvals=nullvalues)
```

```python
# Read data and drop redundant cols
df = pd.read_csv(file_name,skipinitialspace=True,sep=';')
df.drop('education-id',1, inplace=True)
df[continuous_cols] = zscore(df[continuous_cols])
```

```python
# Make categorical columns continuous
orders = {}
for c in df[categorical_cols].columns:
    tmp = df.pivot_table(aggfunc=len, index=c, columns=class_col, values=[1], fill_value=0)
    tmp = tmp / tmp.sum()
    # Add labels
    tmp['ratio'] = tmp[1]/tmp[0]
    tmp = tmp.drop([0,1],1).sort(columns='ratio', ascending=False)
    tmp['value']=tmp.index
    tmp = pd.DataFrame(tmp['value'])
    tmp['i'] = index=range(tmp.shape[0])
    orders[c]=dict(np.array(tmp))
for d in orders:
    df[d] = df[d].map(orders[d]).map(int)
```

```python
# Check new distribution og class frequencies on categorical variables
tools.plot_class_frequency_distribution(df, categorical_cols, class_col, figcols=3)
```

```python
tools.plot_correlation_heatmap(df)
```

```python
X, y = df.drop(class_col,1), df[class_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
models = {
    'GradientBoosting' : GradientBoostingClassifier(n_estimators=200, max_depth=5),
    'RandomForest' : RandomForestClassifier(),
    'SGD' : SGDClassifier(loss='log'),
    'LogReg' : LogisticRegression(),
    'AdaBoost' : AdaBoostClassifier(n_estimators=300, learning_rate=0.05),
    #'DecisionTree' : DecisionTreeClassifier()
}
for m in models.values():
    m.fit(X_train, y_train)

fig, axes = plt.subplots(1)
legend =[]
for m in models.keys():
    predictions = models[m].predict_proba(X_test)[:,1]
    tpr, fpr, thresholds = roc_curve(y_test, predictions)
    axes.plot(tpr, fpr)
    legend.append(m+" ("+str(round(roc_auc_score(y_test, predictions)*100,1))+")")
axes.legend(legend, loc='lower right')
axes.grid()
fig.set_size_inches(12,6)
```

```python
# Show feature importances
feat_imps = pd.DataFrame()
for m in models:
    try:
        feat_imps[m] = models[m].feature_importances_
        feat_imps.index = X.columns
    except AttributeError:
        print m + ' does not have feature importances.'

fig, ax = plt.subplots()
fig.set_size_inches(8,8)
_ = feat_imps.plot(kind='barh', alpha=0.7, ax=ax)
```

```python
features = [0,1,2,3,5,4,6,7,8,9,10,11]
model = 'GradientBoosting'
fig, ax = plt.subplots()
fig.set_size_inches(16,14)
fig, axs = partial_dependence.plot_partial_dependence(models[model], X,
                                                      features, feature_names=X.columns, ax=ax, n_jobs=-1)
```

```python
# X-validate models
scores = {}
for m in models:
    score = tools.cross_val_score(models[m],X,y, cv=4, scoring='roc_auc', n_jobs=-1)
    scores[m] = score.mean()
scores
```
